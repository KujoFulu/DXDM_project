### Plots of the simulation results

```python
# Import packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#### 1. The data plots from original paper

```python
# Load the provided CSV file
file_path = '/Users/kujofulu/go/src/DXDM_project/LotkaVolterra/output/data_paper.csv'
data = pd.read_csv(file_path)

# Plotting the data
plt.figure(figsize=(12, 6))

# Plot each species as a separate line
plt.plot(data['Generation'], data['Species 0'], label='Species 0', color='blue')
plt.plot(data['Generation'], data['Species 1'], label='Species 1', color='green')
plt.plot(data['Generation'], data['Species 2'], label='Species 2', color='red')

# Adding title and labels
plt.title('Species Population Over Generations')
plt.xlabel('Generation')
plt.ylabel('Population')
plt.legend()

# Set export path
export_file_path = "/Users/kujofulu/go/src/DXDM_project/LotkaVolterra/output/data_paper.png"

# Export the plot as a png file
plt.savefig(export_file_path)

# Show the plot
plt.show()
```

#### 6. Draw the real data set: lynx - hare

```python
# Define data
file_path_lynx_hare = "/Users/kujofulu/go/src/DXDM_project/LotkaVolterra/realdata/hudson_bay_lynx_hare.csv"
data_lynx_hare = pd.read_csv(file_path_lynx_hare, skiprows=2)

# Read data
data_lynx_hare.columns = data_lynx_hare.columns.str.strip()

# Plot the data
plt.figure(figsize=(12, 6))

# Plotting Lynx and Hare populations over the years
plt.plot(data_lynx_hare['Year'], data_lynx_hare['Lynx'], label='Lynx', color='blue')
plt.plot(data_lynx_hare['Year'], data_lynx_hare['Hare'], label='Hare', color='red')

# Adding title and labels
plt.title('Hudson Bay Lynx and Hare Populations Over Time')
plt.xlabel('Year')
plt.ylabel('Population (in thousands)')
plt.legend()

# Set export path
export_file_path = "/Users/kujofulu/go/src/DXDM_project/LotkaVolterra/output/hudson_bay_lynx_hare.png"

# Export the plot as a png file
plt.savefig(export_file_path)

# Show the plot
plt.show()
```
