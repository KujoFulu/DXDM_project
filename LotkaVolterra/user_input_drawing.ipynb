#### Function for drawing multiple species simulation output

Note: For the simulation operated in Rshiny, I changed the main fuction to set all out put files' name as "test.csv" and "test.out.gif". So, to use this function for generating user-input simulations, just set the path according to these two file names.

```python
# Import packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the provided CSV file
file_path = '/Users/kujofulu/go/src/DXDM_project/LotkaVolterra/output/test.csv'
data = pd.read_csv(file_path)

# Function to plot different numbers of species with random colors
def plot_species(data):
    plt.figure(figsize=(12, 6))

    # Getting the number of species columns (excluding 'Generation' column)
    num_species = len(data.columns) - 1
    colors = plt.cm.jet(np.linspace(0, 1, num_species))

    # Plot each species as a separate line with a unique color
    for i in range(num_species):
        plt.plot(data['Generation'], data[data.columns[i+1]], label=f'Species {i}', color=colors[i])

    # Adding title and labels
    plt.title('Species Population Over Generations')
    plt.xlabel('Generation')
    plt.ylabel('Population')
    plt.legend()
    
    # Set export path
    export_file_path = "/Users/kujofulu/go/src/DXDM_project/LotkaVolterra/output/test.png"

    # Export the plot as a png file
    plt.savefig(export_file_path)

    # Show the plot
    plt.show()

# Calling the function with the loaded data
plot_species(data)
```
